```python
from transformers import pipeline
from langchain_core.prompts import PromptTemplate
from langchain_core.runnables import RunnableLambda, RunnablePassthrough

# Load summarization model locally
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

# LangChain-compatible wrapper
def summarize_text(input: dict) -> str:
    summary = summarizer(input["text"], max_length=130, min_length=30, do_sample=False)
    return summary[0]["summary_text"]

# LangChain flow
chain = RunnablePassthrough.assign(summary=RunnableLambda(summarize_text))

# Run it
if __name__ == "__main__":
    text = input("📝 Enter text to summarize: ")
    result = chain.invoke({"text": text})
    print("\n📚 Summary:\n", result["summary"])
```
