```python
#Suggestions/Insights Using FLAN-T5 (Second LLM)
from transformers import pipeline
from langchain_core.runnables import RunnableLambda

# 🔹 Load local text2text generation model
commentary_pipeline = pipeline("text2text-generation", model="google/flan-t5-base")

# 🔹 Define LangChain-style function
def generate_commentary(inputs: dict) -> str:
    summary = inputs["summary"]
    prompt = f"Based on this summary, provide suggestions or improvements:\n\n{summary}"
    result = commentary_pipeline(prompt, max_new_tokens=100)
    return result[0]["generated_text"]

# 🔹 LangChain wrapper
commentary_chain = RunnableLambda(generate_commentary)

# 🧪 Run independently (optional)
if __name__ == "__main__":
    summary_input = input("Enter summary text to generate suggestions: ")
    result = commentary_chain.invoke({"summary": summary_input})
    print("\n💡 Suggestions:\n", result)
```
