```python
from transformers import pipeline

# 🔄 Load the BART summarization model from Hugging Face
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

def summarize_news(article_text):
    # 📏 BART has a 1024-token input limit. Use chunks to handle longer text.
    max_chunk = 1024

    # ✂️ Split text into manageable chunks
    article_chunks = [article_text[i:i+max_chunk] for i in range(0, len(article_text), max_chunk)]

    summary = ""

    for chunk in article_chunks:
        result = summarizer(chunk, max_length=130, min_length=30, do_sample=False)
        summary += result[0]['summary_text'] + " "

    return summary.strip()

# ✏️ Main input/output
if __name__ == "__main__":
    print("📝 Enter the article text you want to summarize:")
    article = input()

    print("\n=== 🧠 Summary ===")
    print(summarize_news(article))
```
