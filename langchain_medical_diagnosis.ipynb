```python
from transformers import pipeline
from langchain_core.prompts import PromptTemplate
from langchain_core.runnables import RunnableLambda, RunnablePassthrough

# Step 1: Load a local model that can handle text generation or question answering
generator = pipeline("text-generation", model="gpt2")  # You can use any local model

# Step 2: Create a custom prompt template
prompt = PromptTemplate(
    input_variables=["symptoms"],
    template="""
You are a medical assistant AI. Based on the symptoms described, provide possible diagnoses or advice.

Symptoms: {symptoms}

Diagnosis or Recommendation:
"""
)

# Step 3: Wrap the model in LangChain logic
def diagnose(input: dict) -> str:
    prompt_text = prompt.format(symptoms=input["symptoms"])
    response = generator(prompt_text, max_length=150, do_sample=False)
    return response[0]["generated_text"].replace(prompt_text, "").strip()

# Step 4: Build the chain
chain = RunnablePassthrough.assign(diagnosis=RunnableLambda(diagnose))

# Step 5: Run the assistant locally
if __name__ == "__main__":
    user_input = input("🩺 Describe the symptoms: ")
    result = chain.invoke({"symptoms": user_input})
    print("\n🧠 Possible Diagnosis:\n", result["diagnosis"])
```
